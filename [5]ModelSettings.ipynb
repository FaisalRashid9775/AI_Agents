# **Model Setting** :

model_setting provide you more customize setting to your agent. You can set temperature max_tokens and tool_choice.

1. Temperature : set creativity of the LLM. 0 for less creativity and 2 for maximum creativity
2. max_tokens : set maximium number of tokens
3. tool_choice : Give customizationa about tool.

```python
from agents import Runner, Agent, AsyncOpenAI,set_tracing_disabled,OpenAIChatCompletionsModel,function_tool, ModelSettings
from agents.run import RunConfig
from google.colab import userdata
api_key= userdata.get('GOOGLE_API_KEY')

import nest_asyncio
nest_asyncio.apply()

client : AsyncOpenAI = AsyncOpenAI(
    api_key=api_key,
    base_url="https://generativelanguage.googleapis.com/v1beta/openai/")

model : OpenAIChatCompletionsModel = OpenAIChatCompletionsModel(
    model = 'gemini-2.5-flash',
    openai_client=client
)

set_tracing_disabled(disabled= True)

@function_tool()
def add (a:int,b:int) -> int:
  print('Tool Add Calling')
  return a+b

math_agent = Agent(
    name='Math-agent',
    instructions='You are helpful math agent',
    model=model,
    tools = [add],
    model_settings=ModelSettings(
        temperature=1,
        max_tokens=10000
        )
    )
result = Runner.run_sync(math_agent,'what is 2+2')
print(result.final_output)
```

# ***Tool_choice*** :
 It control the tool call behavior of LLM. It provide three tool_choice.
1. auto : default setting it call tool acording to its choice.
2. none : this setting does not call tool.
3. required : this setting must call tool.

```python
math_agent = Agent(
    name='Math-agent',
    instructions='You are helpful math agent',
    model=model,
    tools = [add],
    model_settings=ModelSettings(
        # tool_choice='none',
        tool_choice = 'required',

        )
    )
result = Runner.run_sync(math_agent,'what is 2+2')
print(result.final_output)
```
