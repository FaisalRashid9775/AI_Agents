# **Agent Handoffs**  

A agents can handoff whole task to another specialize agent. But it not only handoff task but also handoff the complete controls.

```python
from agents import Agent,Runner,OpenAIChatCompletionsModel,AsyncOpenAI,set_tracing_disabled, function_tool, RunContextWrapper
import os
from google.colab import userdata
api_key= userdata.get('GOOGLE_API_KEY')
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
client : AsyncOpenAI = AsyncOpenAI(
    api_key=api_key,
    base_url="https://generativelanguage.googleapis.com/v1beta/openai/")

lite_model : OpenAIChatCompletionsModel = OpenAIChatCompletionsModel(
    model = 'gemini-1.5-flash',
    openai_client=client
)
pro_model : OpenAIChatCompletionsModel = OpenAIChatCompletionsModel(
    model = 'gemini-2.5-flash',
    openai_client=client
)
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")

import nest_asyncio
nest_asyncio.apply()
```

```python
math_agent = Agent(
    name='math-agent',
    instructions='''You are helpful agent.
    Your main responsibility to solve Math Question
    ''',
    model=pro_model
    )

physics_agent = Agent(
    name='physics-agent',
    instructions='''You are helpful agent
     Your main responsibility to answer query Physics related only.
    ''',
    model=pro_model
    )
computer_Agent = math_agent.clone(
    name = 'computer-agent',
    instructions = 'You are a helpful computer agent'

)
chemistry_agent = Agent(
    name='chemistry-agent',
    instructions='''You are helpful agentYou are helpful agent Your main responsibility to answer query chemistry related only. ''',
    model=lite_model,

    )

main_agent = Agent(
    name='main-agent',
    instructions='''You are helpful agent.You are only allowed to Answer the question related to math, physic, chemistry related. Check this before
    # answering the question.If user ask about physics question handoffs to physics agent. If user ask about math question handoffs to math agent. If user ask
    # about chemistry question handoffs to chemistry agent was handoffs the task to whom. Always tell the user which agent.
      ''',
    model=pro_model,
    handoffs = [math_agent,physics_agent,chemistry_agent,computer_Agent], #It handoffs to these four agents.

    )


reponse  = Runner.run_sync(main_agent,' what is binary code')

print(reponse.final_output)
print(f'Replaing Agent : {reponse.last_agent.name}')


```

# **Agent as Tool:**

We can pass an agent as tool to another agent. In this main agent only call other agent as tool. It will not give completes control of task.

```python
math_agent = Agent(
    name='math-agent',
    instructions='''You are helpful agent.
    Your main responsibility to solve Math Question
    ''',
    model=pro_model
    )

physics_agent = Agent(
    name='physics-agent',
    instructions='''You are helpful agent
     Your main responsibility to answer query Physics related only.
    ''',
    model=pro_model
    )
computer_Agent = math_agent.clone(
    name = 'computer-agent',
    instructions = 'You are a helpful computer agent'

)
chemistry_agent = Agent(
    name='chemistry-agent',
    instructions='''You are helpful agentYou are helpful agent Your main responsibility to answer query chemistry related only. ''',
    model=lite_model,

    )

main_agent = Agent(
    name='main-agent',
    instructions='''You are helpful agent.You are only allowed to Answer the question related to math, physic, chemistry related. Check this before
    # answering the question.If user ask about physics question handoffs to physics agent. If user ask about math question handoffs to math agent. If user ask
    # about chemistry question handoffs to chemistry agent was handoffs the task to whom. Always tell the user which agent.
      ''',
    model=pro_model,
    tools = [math_agent.as_tool(
        tool_name='math_agent',
        tool_description='use for math related questions'
    ),
    physics_agent.as_tool(
        tool_name='physics_agent',
        tool_description='use for physics related questions'
    ),chemistry_agent.as_tool(
        tool_name='chemistry_agent',
        tool_description='use for chemistry related questions'
    ),computer_Agent.as_tool(
        tool_name='computer_agent',
        tool_description='use for computer related questions'
    )

    ]


    )


reponse  = Runner.run_sync(main_agent,' what is binary code')

print(reponse.final_output)
print(f'Replaing Agent : {reponse.last_agent.name}')
```
