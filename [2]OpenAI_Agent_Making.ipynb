# **OpenAI Agent SDK **

Open AI Agent SDK is a frame work. It provide ease while development. You did not need to do lengthy code.   


Agent : Agent is AI Software that can take action independly.It Consist two thing an LLM (Brain) and a tool (simple python function) use to perform an action. Agent are autonomous and can do action. Normally it do repetitive task

```python
from agents import Runner, Agent, AsyncOpenAI,set_tracing_disabled,OpenAIChatCompletionsModel
from google.colab import userdata

# Getting API Key
api_key= userdata.get('GOOGLE_API_KEY')

# Colab is already in a Loop So we Call this
import nest_asyncio
nest_asyncio.apply()

#Starter Code/ Boiler Plate Use when you dont have OpenAI Keys.

#Making Client
client : AsyncOpenAI = AsyncOpenAI(
    api_key=api_key,
    base_url="https://generativelanguage.googleapis.com/v1beta/openai/")

#Setting Model
model : OpenAIChatCompletionsModel = OpenAIChatCompletionsModel(
    model = 'gemini-2.5-flash',
    openai_client=client
)
set_tracing_disabled(disabled=True)

#Making Agents
math_agent = Agent(name='Math-agent',instructions='You are helpful math agent',model=model)
result = Runner.run_sync(math_agent,'what is 2+2')
print(result.final_output)
```

run_sync run agent synchronously.But you can run asynchronously by running this code.

# Making it Async

```python
import asyncio
async def main():
  math_agent = Agent(name='Math-agent',instructions='You are helpful math agent',model=model)
  result = await Runner.run(math_agent,'what is 2+2')
  print(result.final_output)
asyncio.run(main())
```

# **Streaming**

```python
from openai.types.responses import ResponseTextDeltaEvent
import asyncio
async def main():
  math_agent = Agent(name='Math-agent',instructions='You are helpful math agent',model=model)
  result = Runner.run_streamed(math_agent,'what is 2+2 explain with examples')
  async for event in result.stream_events():
            if event.type == "raw_response_event" and isinstance(event.data, ResponseTextDeltaEvent):
                print('AGENT : ',event.data.delta, end="", flush=True)

asyncio.run(main())
```
